```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as seabornInstance 
df = pd.read_csv('Chile.csv')
x = df['Month']
y = df['Users']
y.describe()
```

```python
seabornInstance.set_style(style='whitegrid')
plt.rcParams["patch.force_edgecolor"] = True
```

```python
fig, ax = plt.subplots(figsize=(12,8))

plt.barh(x, y, color = ['bisque', 'darkgray', 'darkgray','darkgray','black','darkgray'])
for index, value in enumerate(y):
    plt.text(value, index, str(value))
    
plt.title('Users in Chile starting Jan 2020 to mid June 2020')
ax.set_xlabel('Users')
ax.set_ylabel('Month')
y_label = ('Jan', 'Feb', 'March', 'April', 'May', 'June')
ax.set_yticklabels(y_label)

```
